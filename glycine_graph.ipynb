## График с IR всех результатов

- **Глицин** - эксперментальный, немодифицированный
- **Глицин ZW + Вода** - расчетный
- **Глицин ZW** - расчетный
- **2Глицин ZW** - расчетный
- **2Глицин ZW + 4Вода** - расчетный
- **2Глицин ZW + 6Вода** - расчетный

```python
# %matplotlib inline

import matplotlib.pyplot as plt
import pandas as pd

def do_anharmonism_correction(x):
    if 1000 < x < 2000:
        x = x - 1000
        x = (x * 0.98) + 1000
    elif x > 2000:
        x = x - 2000
        x = (x * 0.95) + 2000
    return x


headers1 = ['freq', 'Gly Original']
headers2 = ['freq', 'Gly Modify']
df1 = pd.read_csv('Глицин Fluka.CSV', names=headers1)
df2 = pd.read_csv('zwitterion_solv_ir.txt', names=headers2)
df3 = pd.read_csv('zwitterion_glycine_ir2.txt', names=headers2)
df4 = pd.read_csv('2zw_GLy_2ver_freq_ir.txt', names=headers2)
df5 = pd.read_csv('2ZwGLY_water_freq_ir.txt', names=headers2)
df6 = pd.read_csv('2ZwGLY_6water_freq_ir2.txt', names=headers2)

x1 = df1['freq']
y1 = df1['Gly Original'] * df2['Gly Modify'].max()
x2 = df2['freq'].map(lambda x: do_anharmonism_correction(x))
y2 = df2['Gly Modify']

x3 = df3['freq'].map(lambda x: do_anharmonism_correction(x))
y3 = df3['Gly Modify']*3

x4 = df4['freq'].map(lambda x: do_anharmonism_correction(x))
y4 = df4['Gly Modify']

x5 = df5['freq'].map(lambda x: do_anharmonism_correction(x))
y5 = df5['Gly Modify']

x6 = df6['freq'].map(lambda x: do_anharmonism_correction(x))
y6 = df6['Gly Modify']



plt.plot(x1, y1, lw=1, alpha=1)
plt.plot(x2, y2, lw=1, alpha=1)
plt.plot(x3, y3, lw=1, alpha=1)
plt.plot(x4, y4, lw=1, alpha=1)
plt.plot(x5, y5, lw=1, alpha=1)
plt.plot(x6, y6, lw=1, alpha=1)

plt.legend(['Глицин',  'Глицин ZW + Вода', 'Глицин ZW', '2Глицин ZW',  '2Глицин ZW + 4Вода', '2Глицин ZW + 6Вода'])
plt.rcParams['figure.figsize'] = [20, 10]
plt.rcParams['figure.dpi'] = 200

plt.show()
```

## График сравнения глицина, 2 глицина + 4 вода

- **2 Глицин ZW** - экспериментальный
- **2 Глицин ZW + 4 Вода** - расчетный

```python
# %matplotlib inline

import matplotlib.pyplot as plt
import pandas as pd

def do_anharmonism_correction(x):
    if 1000 < x < 2000:
        x = x - 1000
        x = (x * 0.98) + 1000
    elif x > 2000:
        x = x - 2000
        x = (x * 0.95) + 2000
    return x


headers1 = ['freq', 'Gly Original']
headers2 = ['freq', 'Gly Modify']
df1 = pd.read_csv('Глицин Fluka.CSV', names=headers1)
df5 = pd.read_csv('2ZwGLY_water_freq_ir.txt', names=headers2)


x1 = df1['freq']
y1 = df1['Gly Original'] * df5['Gly Modify'].max()
x5 = df5['freq'].map(lambda x: do_anharmonism_correction(x))
y5 = df5['Gly Modify']

plt.plot(x1, y1, lw=2, alpha=1)
plt.plot(x5, y5, lw=2, alpha=1)


plt.legend(['Глицин экспериментальный', '2Глицин ZW + 4Вода'])
plt.rcParams['figure.figsize'] = [20, 10]
plt.rcParams['figure.dpi'] = 200

plt.show()
```
